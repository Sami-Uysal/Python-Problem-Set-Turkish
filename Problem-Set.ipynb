1)	En az 10 kullanıcının ismini içinde bulunduran bir liste ve boş bir liste olmak üzere iki liste tanımlayın ve daha sonra en az 10 kullanıcının adını içinde bulunduran listeden random modülü yardımıyla sırasıyla üç ismi rastgele seçerek ikinci boş listeye ekleyin daha sonra çekilişi kazananlar diye ikinci listenin içerisindeki isimleri sırasıyla consola yazdırın.

Örnek Çıktı:
Çekilişi kazananlar:
Ahmet
Hasan
Selin


```python
import random
kullanıcılar=[]
kazananlar=[]
kullanıcı=input('isimler(virgüllü yazın):')
for i in kullanıcı.split(','):
    kullanıcılar.append(i)
if len(kullanıcılar)<10:
    print('en az 10')
elif len(kullanıcılar)>=10:
    x=random.choice(kullanıcılar)
    kazananlar.append(x)
    y=random.choice(kullanıcılar)
    while y==x:
        y=random.choice(kullanıcılar)
    else:
        kazananlar.append(y)
    z=random.choice(kullanıcılar)
    while z==y or z==x:
        z=random.choice(kullanıcılar)
    else:
        kazananlar.append(z)
for kazanan in kazananlar:
    print(kazanan)
```

2)	Sayı tahmin oyunu yapalım, 1 ile 100 arasında rastgele bir sayı belirleyip kullanıcının bu sayıyı tahmin etmesini sağlayalım kullanıcı sayıyı doğru bilinceye kadar sürekli olarak bir döngü içerisinde kullanıcıdan sayıyı alalım ve eğer kullanıcının girdiği sayı rastgele üretilen sayıdan büyükse program kullanıcıya daha küçük bir sayı girin diye bilgi versin aynı şekilde kullanıcının girdiği sayı rastgele üretilen sayıdan küçükse program kullanıcıya daha büyük bir sayı girmesi konusunda bilgi versin kullanıcının sayıyı doğru tahmin etmesi durumunda ise program tebrikler sayıyı doğru tahmin ettiniz tarzında bir bilgi vererek döngüyü sonlandırsın.

```python
import random
num=random.randint(0,101)
sayi=int(input('bir sayi girin:'))
if sayi==num:
    print('tebrikler bildiniz')
else:
    while sayi!=num:
        if sayi>num:
            print('daha küçük')
        elif sayi<num:
            print('daha büyük')
        sayi=int(input('tekrar deneyiniz:'))
        if sayi==num:
            print('tebrikler')
```

Örnek Çıktı:
Yapmak istediğiniz işlem nedir (1-toplama | 2-çıkarma | 3-bölme | 4-çarpma | 5-çıkış): 
İlk sayıyı giriniz:

```python
def toplama(ilksayi,ikincisayi):
    print(ilksayi+ikincisayi)
def çıkarma(ilksayi,ikincisayi):
    print(ilksayi-ikincisayi)
def bölme(ilksayi,ikincisayi):
    print(ilksayi/ikincisayi)
def çarpma(ilksayi,ikincisayi):
    print(ilksayi*ikincisayi)
while True:
        print("q=çıkış toplama=+ çıkarma=- bölme=÷ çarpma=×")
        islem=input('hangi işlem:')
        if islem=='q':
            break
        ilksayi=int(input('ilksayi:'))
        ikincisayi=int(input('ikincisayi:'))
        if islem=='toplama':
            toplama(ilksayi,ikincisayi)
        elif islem=='çıkarma':
            çıkarma(ilksayi,ikincisayi)
        elif islem=='bölme':
            bölme(ilksayi,ikincisayi)
        elif islem=='çarpma':
            çarpma(ilksayi,ikincisayi)
        else:
            print('geçerli işlem girin')
```

```python
boy=int(input('boyunuz(santimetre cinsinden):'))
kilo=int(input('kilonuz(kg cinsinden):'))
endeks=kilo/(boy/100)**2
print(endeks)
if endeks<=18.5:
    print('Zayıf')
elif endeks>18.5 and endeks<=25:
    print('Normal')
elif endeks>25 and endeks<=30:
    print('Kilolu')
elif endeks>30:
    print('Obez')
```

5)	Bir fonksiyon tanımlayın bu fonksiyon kendisine parametre olarak verilen değerin 10 tabanında logaritmasını, karekökünü ve mutlak değerini hesaplayarak sonuclar.txt diye bir dosyaya alt alta yazdırsın. (sonuçlar.txt dosyasını programın çalıştığı python dosyasıyla aynı dizinde olsun ve eğer ki sonuclar.txt adında bir dosya mevcut dizinde varsa o dosyanın içerisine yazdırma işlemi yapılsın aksi taktirde dizinde sonuclar.txt adında bir dosya oluşturulsun ve sonuçlar içerisine yazılsın.)

```python
import math
sayi=int(input('sayi girin:'))
file=open('dosya konumu','w')
logaritma=math.log(sayi,10)
karekok=sayi*0.5
if sayi<0:
    mutlak=sayi*-1
else:
    mutlak=sayi
file.write('{0}\n{1}\n{2}'.format(logaritma,karekok,mutlak))

file.close()
```
